```python
class Node:
    def __init__(self,data,prev,next):
        self.data = data
        self.prev = prev
        self.next = next
```

```python
class Doublylinkedlist():
    def __init__(self):
        self.head = None
    def print(self):
        pass
        if self.head == None:
            return("empty LL")
        else:
            llstr = ''
            itr = self.head
            while itr:
                llstr += str(itr.data) + '<-->'
                itr = itr.next
            print(llstr)
            return(llstr)

    def get_length(self):
        pass
        if self.head == None:
            print("empty ll")
            return("Empty LL")
        else:
            count = 0
            itr = self.head
            while itr:
                count +=1
                itr = itr.next
            print(count)
            return count

    def insert_at_begining(self, data):
        if self.head == None:
            node = Node(data,None,self.head)
            self.head = node
        else:
            node = Node(data,None,self.head)
            self.head.prev = node
            self.head = node

    def insert_at_end(self, data):
        if self.head ==  None:
            self.insert_at_begining(data)
            return
        else:
            itr = self.head
            while itr:
                if itr.next == None:
                    node = Node(data,itr,None)
                    itr.next = node
                    break
                itr = itr.next

    def insert_at(self, index, data):
        pass
        if index < 0 or index > self.get_length():
            print("invalid index")
            return
        else:
            count = 0
            itr = self.head
            while itr:
                if count == index -1:
                    node = Node(data,itr,itr.next)
                    itr.next.prev = node
                    itr.next = node
                    
                    break
                count += 1
                itr = itr.next
    def remove_at(self, index):
        if index < 0 or index > self.get_length():
            print("invalid index")
            return
        else:
            count = 0
            itr = self.head
            while itr:
                if count == index -1:
                    itr.next.prev = itr
                    itr.next = itr.next.next
                    break
                count += 1
                itr = itr.next
        

    def insert_values(self, data_list):
        for data in data_list:
            self.insert_at_end(data)
        return
    
    
    def insert_after_value(self, data_after, data_to_insert):
        
        itr =  self.head
        while itr:
            if itr.data == data_after:
                node = Node(data_to_insert,itr,itr.next)
                itr.next.prev = node
                itr.next = node
                break
            itr = itr.next
        return

    def remove_by_value(self, data):
        
        if self.head == None:
            return
        if self.head.data == data:
            self.head = self.head.next
            return
        itr  = self.head
        while itr.next:
            if itr.next.data == data:
                itr.next.prev = itr
                itr.next = itr.next.next
                return
            itr =itr.next
        return(print("invalid value"))
    
    def print_backward(self):
        if self.head == None:
            print("empty LL")
            return
        itr = self.head
        llstr = ""
        while itr:
            if itr.next == None:
                while itr:
                    llstr += str(itr.data)  + '-->'
                    itr = itr.prev
                print(llstr)
                return
            itr = itr.next
```

```python
if __name__ == '__main__':
    ll = Doublylinkedlist()
    ll.insert_at_begining(11)
    ll.insert_at_begining(22)
    ll.insert_at_begining(33)
    ll.insert_values(["banana","mango","grapes","orange"])
    ll.insert_at(1,"blueberry")
    ll.insert_at_end(67)
    ll.remove_at(2)
    ll.insert_after_value("mango","apple")
    ll.print()
    ll.print()
    ll.print_backward()
```
